```python
import pandas as pd
from scipy.stats import ttest_ind

def website_analyse(file_path):
    df = pd.read_csv(file_path)

    # 1) Среднее количество просмотренных страниц
    mean_pages = df['Просмотренные страницы'].mean()
    print(f"1) Среднее количество просмотренных страниц: {mean_pages:.2f}")

    # 2) Стандартное отклонение времени на сайте
    std_time = df['Время на сайте (сек)'].std()
    print(f"2) Стандартное отклонение: {std_time:.2f}")

    # 3) Коэффициент вариации
    cv = std_time / df['Время на сайте (сек)'].mean()
    if cv > 0.25:
        variability = "Высокая вариабельность"
    elif cv > 0.1:
        variability = "Умеренная вариабельность"
    else:
        variability = "Слабая вариабельность"
    print(f"3) {variability}: {cv:.3f}")

    # 4) T-тест для мобильных устройств и десктопов
    mobile_pages = df[df['Тип устройства'] == 'Мобильное']['Просмотренные страницы']
    desktop_pages = df[df['Тип устройства'] == 'Десктоп']['Просмотренные страницы']
    t_stat, p_value = ttest_ind(mobile_pages, desktop_pages)
    if p_value > 0.05:
        diff_msg = "Есть статистически значимая разница в количестве страниц"
    else:
        diff_msg = "Нет статистически значимой разницы"
    print(f"4) {diff_msg}: {p_value:.3f}")
```
